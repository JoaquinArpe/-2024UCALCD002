```python
from datetime import datetime

class AdministradorSalud:
    def __init__(self):
        self.registro_ejercicio = []
        self.registro_dieta = []
        self.registro_sueno = []

    def registrar_ejercicio(self, fecha, duracion):
        self.registro_ejercicio.append({"fecha": fecha, "duracion": duracion})

    def registrar_dieta(self, fecha, comidas):
        self.registro_dieta.append({"fecha": fecha, "comidas": comidas})

    def registrar_sueno(self, fecha, horas):
        self.registro_sueno.append({"fecha": fecha, "horas": horas})

    def obtener_registros_ejercicio(self):
        return self.registro_ejercicio

    def obtener_registros_dieta(self):
        return self.registro_dieta

    def obtener_registros_sueno(self):
        return self.registro_sueno

    def calcular_calorias_quemadas(self, duracion):
        # Usaremos una estimación básica de calorías quemadas por minuto para ejemplos
        calorias_por_minuto = 8  # Estimación promedio de calorías quemadas por minuto
        return duracion * calorias_por_minuto

    def calcular_calorias_consumidas(self, comidas):
        # Base de datos de calorías por porción para diferentes tipos de alimentos (simplificado)
        calorias_por_alimento = {
            "desayuno": 300,
            "almuerzo": 500,
            "cena": 400,
            "snack": 150
            # Agrega más alimentos y calorías según sea necesario
        }

        total_calorias = 0
        for comida in comidas:
            comida_lower = comida.lower().strip()  # Convertimos a minúsculas y eliminamos espacios en blanco
            if comida_lower in calorias_por_alimento:
                total_calorias += calorias_por_alimento[comida_lower]
            else:
                print(f"Advertencia: No se encontraron calorías para '{comida}'.")
        return total_calorias


class InterfazUsuarioSalud:
    def __init__(self):
        self.administrador_salud = AdministradorSalud()

    def validar_fecha(self, fecha):
        try:
            datetime.strptime(fecha, '%Y-%m-%d')
            return True
        except ValueError:
            print("Error: La fecha debe estar en el formato AAAA-MM-DD.")
            return False

from datetime import datetime

class AdministradorSalud:
    def __init__(self):
        self.registro_ejercicio = []
        self.registro_dieta = []
        self.registro_sueno = []

    def registrar_ejercicio(self, fecha, duracion):
        self.registro_ejercicio.append({"fecha": fecha, "duracion": duracion})

    def registrar_dieta(self, fecha, comidas):
        self.registro_dieta.append({"fecha": fecha, "comidas": comidas})

    def registrar_sueno(self, fecha, horas):
        self.registro_sueno.append({"fecha": fecha, "horas": horas})

    def obtener_registros_ejercicio(self):
        return self.registro_ejercicio

    def obtener_registros_dieta(self):
        return self.registro_dieta

    def obtener_registros_sueno(self):
        return self.registro_sueno

    def calcular_calorias_quemadas(self, duracion):
        # Usaremos una estimación básica de calorías quemadas por minuto para ejemplos
        calorias_por_minuto = 8  # Estimación promedio de calorías quemadas por minuto
        return duracion * calorias_por_minuto

    def calcular_calorias_consumidas(self, comidas):
        # Base de datos de calorías por porción para diferentes tipos de alimentos (simplificado)
        calorias_por_alimento = {
            "desayuno": 300,
            "almuerzo": 500,
            "cena": 400,
            "snack": 150
            # Agrega más alimentos y calorías según sea necesario
        }

        total_calorias = 0
        for comida in comidas:
            comida_lower = comida.lower().strip()  # Convertimos a minúsculas y eliminamos espacios en blanco
            if comida_lower in calorias_por_alimento:
                total_calorias += calorias_por_alimento[comida_lower]
            else:
                print(f"Advertencia: No se encontraron calorías para '{comida}'.")
        return total_calorias


class InterfazUsuarioSalud:
    def __init__(self):
        self.administrador_salud = AdministradorSalud()

    def validar_fecha(self, fecha):
        try:
            datetime.strptime(fecha, '%Y-%m-%d')
            return True
        except ValueError:
            print("Error: La fecha debe estar en el formato AAAA-MM-DD.")
            return False

    def validar_comidas(self, comidas):
        alimentos_conocidos = ["desayuno", "almuerzo", "cena", "snack"]
        for comida in comidas:
            if comida.lower().strip() not in alimentos_conocidos:
                print(f"Error: '{comida}' no es un tipo de comida válido.")
                return False
        return True

    def registrar_ejercicio(self):
        fecha = input("Ingrese la fecha (AAAA-MM-DD): ")
        duracion = int(input("Ingrese la duración del ejercicio (minutos): "))
        if duracion> 300:
          print("--------------------------------------------")
          print("Cuidado pa, no es salubre")
          print("--------------------------------------------")
        self.administrador_salud.registrar_ejercicio(fecha, duracion)
        calorias_quemadas = self.administrador_salud.calcular_calorias_quemadas(duracion)
        print(f"Registro de ejercicio exitoso. Calorías quemadas aproximadas: {calorias_quemadas}")

    def registrar_dieta(self):
        fecha = input("Ingrese la fecha (AAAA-MM-DD): ")
        if not self.validar_fecha(fecha):
            return

        comidas = input("Ingrese las comidas separadas por comas: ").split(',')
        if not self.validar_comidas(comidas):
            return

        total_calorias = self.administrador_salud.calcular_calorias_consumidas(comidas)
        self.administrador_salud.registrar_dieta(fecha, comidas)
        print(f"Registro de dieta exitoso. Total de calorías consumidas: {total_calorias}")

    def registrar_sueno(self):
        fecha = input("Ingrese la fecha (AAAA-MM-DD): ")
        horas = int(input("Ingrese las horas de sueño: "))
        self.administrador_salud.registrar_sueno(fecha, horas)
        print("Registro de sueño exitoso.")

    def mostrar_registros(self):
        print("\nRegistros de Ejercicio:")
        for entrada in self.administrador_salud.obtener_registros_ejercicio():
            fecha = entrada['fecha']
            duracion = entrada['duracion']
            calorias_quemadas = self.administrador_salud.calcular_calorias_quemadas(duracion)
            print(f"Fecha: {fecha}, Duración: {duracion} minutos, Calorías quemadas: {calorias_quemadas}")

        print("\nRegistros de Dieta:")
        for entrada in self.administrador_salud.obtener_registros_dieta():
            fecha = entrada['fecha']
            comidas = ', '.join(entrada['comidas'])
            total_calorias = self.administrador_salud.calcular_calorias_consumidas(entrada['comidas'])
            print(f"Fecha: {fecha}, Comidas: {comidas}, Total de calorías consumidas: {total_calorias}")

    def ejecutar(self):
        while True:
            print("\nBienvenido a la Aplicación de Gestión de Salud y Bienestar")
            print("1. Registrar Ejercicio")
            print("2. Registrar Dieta")
            print("3. Registrar Sueño")
            print("4. Ver Registros")
            print("5. Salir")

            opcion = input("Seleccione una opción: ")

            if opcion == "1":
                self.registrar_ejercicio()
            elif opcion == "2":
                self.registrar_dieta()
            elif opcion == "3":
                self.registrar_sueno()
            elif opcion == "4":
                self.mostrar_registros()
            elif opcion == "5":
                print("¡Hasta luego!")
                break
            else:
                print("Opción inválida. Por favor, seleccione una opción válida.")

# Creamos una instancia de la interfaz de usuario y ejecutamos la aplicación
if __name__ == "__main__":
    interfaz_usuario_salud = InterfazUsuarioSalud()
    interfaz_usuario_salud.ejecutar()







```
